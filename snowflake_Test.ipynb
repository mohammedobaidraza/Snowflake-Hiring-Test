```python

MOD = 10**9 + 7
def numWays(words, target):
    m = len(words[0])
    n = len(target)

    # For loop to create frequency table for each column in words
    freq = [[0] * 26 for _ in range(m)]
    for word in words:
        for i in range(m):
            freq[i][ord(word[i]) - ord('a')] += 1

    # Function to initialize DP array for ways to build target string
    dp = [0] * (n + 1)
    dp[0] = 1

    # For loop to update DP array using frequency data
    for i in range(m):
        for j in range(n - 1, -1, -1):
            char_index = ord(target[j]) - ord('a')
            dp[j + 1] = (dp[j + 1] + dp[j] * freq[i][char_index]) % MOD

    return dp[n]

# Using main function to check inputs and output
if __name__ == '__main__':
    words_count = int(input().strip())
    words = [input().strip() for _ in range(words_count)]
    target = input().strip()

    result = numWays(words, target)
    print(result)
```
